```python
classlabelvector = df.ix[:, 3]
classlabelvector.head()
```

```python
datinglabels = np.array(classlabelvector)
datinglabels[0:20]
```

```python
ll_datainglabels = []
for i in datinglabels:
    if i == 'largeDoses':
        ll_datainglabels.append(3)
    elif i == 'smallDoses':
        ll_datainglabels.append(2)
    else:
        ll_datainglabels.append(1)
ll_datainglabels
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(111)
```

```python
ax.scatter(dataingdatamat[:, 1],dataingdatamat[:, 2], 
           15.0*np.array(ll_datainglabels), 15.0*np.array(ll_datainglabels))
ax.axis([-2,25,-0.2,2.0])
plt.show()
```
